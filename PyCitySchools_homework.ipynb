```python
import pandas as pd
import numpy as np

School_dSource = "Resources/schools_complete.csv"
Student_dSource = "Resources/students_complete.csv"

SchoolData = pd.read_csv(School_dSource)
StudentData = pd.read_csv(Student_dSource)

SchoolDataMerge = pd.merge(StudentData, SchoolData, how="left", on=["school_name", "school_name"])
SchoolDataMerge.head()

SizeBins = [0, 1000, 2000, 5000]
SizeBinIndex = ["Small (<1K)", "Med (1-2K)", "Large (2-5K)"]

SchoolSizeDF = SchoolDataMerge.groupby(["size"]).mean()[["reading_score","math_score"]]
SchoolSizeDF.reset_index(inplace=True)

SchoolSizeDF.loc[:,'Size Category'] = pd.cut(SchoolSizeDF["size"],SizeBins,labels = SizeBinIndex)

AMS = SchoolSizeDF.groupby(["Size Category"]).mean()["math_score"].round(2)
ARS = SchoolSizeDF.groupby(["Size Category"]).mean()["reading_score"].round(2)

TtlStudentsPerSize = SchoolSizeDF.groupby(["Size Category"]).sum()["size"].round()

PassRead = SchoolDataMerge[SchoolDataMerge['reading_score'] >= 70]
PassRead.loc[:,"Size Category"] = pd.cut(PassRead["size"],SizeBins,labels = SizeBinIndex)
PassReadBins = PassRead.groupby("Size Category").count()["Student ID"]
PercentPassReadPerSize = ((PassReadBins/TtlStudentsPerSize)*100).round(2)

PassMath = SchoolDataMerge[SchoolDataMerge['math_score'] >= 70]
PassMath.loc[:,"Size Category"] = pd.cut(PassMath["size"],SizeBins,labels = SizeBinIndex)
PassMathBins = PassMath.groupby("Size Category").count()["Student ID"]
PercentPassMathPerSize = ((PassMathBins/TtlStudentsPerSize)*100).round(2)

OverallPassPerSize = ((PercentPassReadPerSize + PercentPassMathPerSize)/2).round(2)

SchoolSizeBinnedDict = { 
                        "Avg Math Score": AMS,
                        "Avg Reading Score": ARS,
                        "%Passing Math": PercentPassMathPerSize,
                        "%Passing Reading": PercentPassReadPerSize,
                        "%Passing Overall": OverallPassPerSize
                       }

SchoolSizeBinnedDF = pd.DataFrame(SchoolSizeBinnedDict)

SchoolSizeBinnedDF
```
